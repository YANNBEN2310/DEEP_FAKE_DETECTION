```python
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

On utilise ici `ImageDataGenerator` pour prétraiter les images pendant l'entraînement du modèle. Les paramètres spécifiés incluent :
- `rescale=1./255` : Cela permet de mettre à l'échelle les valeurs de pixel des images pour qu'elles soient comprises entre 0 et 1, ce qui est une bonne pratique pour le traitement des images.
- `shear_range=0.2` : Cela applique un cisaillement aléatoire aux images. Le cisaillement déforme l'image le long d'un axe, ce qui peut être utile pour augmenter la variabilité des données d'entraînement.
- `zoom_range=0.2` : Cela applique un zoom aléatoire aux images, ce qui les agrandit ou les réduit. Cela aide également à augmenter la variabilité des données d'entraînement.
- `horizontal_flip=True` : Cela permet d'effectuer une retournement horizontal aléatoire des images. Cette transformation reflète horizontalement l'image, ce qui peut aider à créer plus de variations dans les données d'entraînement.

```python
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator = datagen.flow_from_directory(
    val_dir,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='categorical'
)
```

On utilise les générateurs de flux (`flow_from_directory`) pour charger les images depuis les répertoires d'entraînement et de validation. Les paramètres spécifiés incluent :
- `train_dir` et `val_dir` : Les chemins vers les répertoires d'entraînement et de validation respectivement.
- `target_size=(150, 150)` : Les images sont redimensionnées pour avoir une taille de 150x150 pixels.
- `batch_size=batch_size` : Le nombre d'images dans chaque lot (32 dans ce cas).
- `class_mode='categorical'` : Les étiquettes sont générées de manière catégorique. Cela signifie que les étiquettes sont représentées sous forme de vecteurs binaires (un vecteur par classe), où une seule composante est définie à 1 (encodage one-hot), indiquant la classe de l'image.

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='sigmoid'))
```

Ce code définit un modèle de réseau de neurones convolutif (CNN) à l'aide de la bibliothèque Keras avec la structure suivante :

- `Sequential()` : Crée un modèle séquentiel, où les couches sont ajoutées les unes après les autres.

- `Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(150, 150, 3))` : Ajoute une couche de convolution avec 32 filtres de taille 3x3, utilisant la fonction d'activation ReLU. La couche prend en entrée des images de taille 150x150 avec 3 canaux (RGB).

- `MaxPooling2D(pool_size=(2, 2))` : Ajoute une couche de max pooling pour réduire la dimensionnalité de la sortie de la couche précédente.

- `Dropout(0.25)` : Ajoute une couche de dropout pour régulariser le modèle en désactivant aléatoirement 25% des neurones pendant l'entraînement, ce qui aide à prévenir le surapprentissage.

- Les mêmes couches sont ensuite répétées, en doublant le nombre de filtres convolutifs à 64.

- `Flatten()` : Ajoute une couche de mise à plat pour convertir les sorties des couches convolutives en un vecteur 1D pour les entrées de la couche dense.

- `Dense(128, activation='relu')` : Ajoute une couche dense avec 128 neurones utilisant la fonction d'activation ReLU.

- `Dropout(0.5)` : Ajoute une autre couche de dropout pour régulariser le modèle en désactivant aléatoirement 50% des neurones.

- `Dense(num_classes, activation='sigmoid')` : Ajoute une couche dense de sortie avec autant de neurones que de classes dans le problème (dans ce cas, 2 classes), utilisant la fonction d'activation sigmoïde pour produire des probabilités pour chaque classe.

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

On compile le modèle défini précédemment. Les paramètres spécifiés dans la méthode `compile` sont les suivants :

- `loss='categorical_crossentropy'` : C'est la fonction de perte utilisée pour évaluer la performance du modèle pendant l'entraînement. Dans ce cas, il s'agit de la perte d'entropie croisée catégorielle, souvent utilisée pour les problèmes de classification avec plusieurs classes.

- `optimizer='adam'` : C'est l'algorithme d'optimisation utilisé pour mettre à jour les poids du réseau pendant l'entraînement. Ici, on utilise l'optimiseur Adam, qui est une méthode d'optimisation populaire et efficace.

- `metrics=['accuracy']` : Cela spécifie les métriques à surveiller pendant l'entraînement. Dans ce cas, la seule métrique spécifiée est l'exactitude (accuracy), qui mesure la proportion d'images correctement classées par le modèle.

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=val_generator,
    validation_steps=val_generator.samples // batch_size
)
```

On entraîne le modèle en utilisant la méthode `fit`, en fournissant les générateurs de données d'entraînement et de validation, ainsi que les paramètres supplémentaires suivants :

- `steps_per_epoch=train_generator.samples // batch_size` : Cela spécifie le nombre total d'étapes (batches) à prendre à chaque époque lors de l'entraînement. Dans ce cas, on divise le nombre total d'échantillons d'entraînement par la taille du lot pour obtenir le nombre d'étapes.

- `epochs=epochs` : Cela spécifie le nombre d'époques (cycles complets sur l'ensemble des données d'entraînement) pour entraîner le modèle. On utilise ici la valeur définie précédemment.

- `validation_data=val_generator` : Cela spécifie les données à utiliser pour l'évaluation de la perte et des métriques à la fin de chaque époque pendant l'entraînement. Dans ce cas, on utilise les données de validation.

- `validation_steps=val_generator.samples // batch_size` : Cela spécifie le nombre total d'étapes (batches) à prendre à chaque époque lors de l'évaluation sur les données de validation. Comme précédemment, on divise le nombre total d'échantillons de validation par la taille du lot pour obtenir le nombre d'étapes.

```python
from sklearn.metrics import classification_report, confusion_matrix

# Generer les predictions sut les validations
y_pred = model.predict(val_generator)
y_pred = np.argmax(y_pred, axis=1)

print(classification_report(val_generator.classes, y_pred))
```

On utilise la fonction `classification_report` de la bibliothèque scikit-learn pour générer un rapport détaillé sur la classification. Ce rapport affiche des mesures telles que la précision, le rappel, le score F1 et le support pour chaque classe dans le jeu de données de validation. Cela fournit une évaluation plus complète de la performance du modèle pour chaque classe.

```python
# Imprimer le rapport de classification
report = classification_report(val_generator.classes, y_pred, output_dict=True)
print(classification_report(val_generator.classes, y_pred))

# Extraire les scores F1 pour chaque classe du rapport de classification
class_names = val_generator.class_indices.keys()
f1_scores = [report[str(i)]['f1-score'] for i, _ in enumerate(class_names)]

# Extraire l'exactitude, la précision et le rappel du rapport de classification
accuracy = report['accuracy']
precision_macro = report['macro avg']['precision']
recall_macro = report['macro avg']['recall']

# Définir un score F1 minimum acceptable
min_f1_score = 0.7

# Déterminer si le modèle est bon ou non en fonction de son score F1
if np.mean(f1_scores) > min_f1_score:
    model_performance = 'bonne'
else:
    model_performance = 'pas bonne'

print(f'Performance du modèle : {model_performance}')

# Tracer les métriques sous forme de graphique à barres
plt.figure(figsize=(10,6))
plt.bar(class_names, f1_scores, color='blue')
plt.xlabel('Classe')
plt.ylabel('Score F1')
plt.title('Scores F1 par Classe')

plt.figure(figsize=(10,6))
metrics_names = ['Exactitude', 'Précision (macro)', 'Rappel (macro)']
metrics_values = [accuracy, precision_macro, recall_macro]
plt.bar(metrics_names, metrics_values, color='green')
plt.xlabel('Métrique')
plt.ylabel('Valeur')
plt.title('Métriques Globales')

plt.show()
```

On imprime d'abord le rapport de classification pour évaluer la performance du modèle pour chaque classe dans le jeu de données de validation. Ensuite, on extrait les scores F1 pour chaque classe du rapport de classification. On extrait également l'exactitude (accuracy), la précision (precision) macro-moyenne et le rappel (recall) macro-moyen du rapport de classification. On définit un score F1 minimum acceptable et on détermine si le modèle est bon ou non en fonction de la moyenne des scores F1. Enfin, on trace les scores F1 pour chaque classe et les métriques globales dans des graphiques à barres pour visualiser la performance du modèle qui ``n'est pas assez performant ``

```python
# Charger le modèle pré-entraîné à partir du fichier h5
model = keras.models.load_model('models/IMAGE_DEEP_FAKE_MODEL-1.h5')

# Définir une fonction pour traiter la sélection des images téléchargées
def select_images(files):
    # Traiter chaque fichier téléchargé
    for file in files:
        # Convertir le fichier téléchargé en un objet BytesIO
        img = image.load_img(io.BytesIO(file['content']), target_size=(150, 150))

        # Convertir l'image en un tableau NumPy et normaliser les valeurs de pixels
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x /= 255.0

        # Faire une prédiction en utilisant le modèle
        y_pred = model.predict(x)

        # Extraire le score de confiance pour la classe prédite
        confidence_score = y_pred[0][1]  # On suppose que le modèle renvoie la probabilité pour la classe '1' (faux)

        # Déterminer l'étiquette et la couleur de l'image en fonction du score de confiance
        if confidence_score >= 0.50:
            label = f'Fake avec un score de confiance de {confidence_score:.2f}'
            color = 'red'
        else:
            label = f'Real avec un score de confiance de {1 - confidence_score:.2f}'  # Le score de confiance pour le réel est 1 - le score pour le faux
            color = 'blue'

        # Afficher l'image avec l'étiquette
        plt.figure()
        plt.imshow(image.load_img(io.BytesIO(file['content'])))
        plt.title(label, color=color)
        plt.axis('off')
        plt.show()

# Définir une fonction pour déclencher les prédictions
def predict_images(button):
    # Obtenir les fichiers téléchargés à partir du widget de sélection de fichiers
    files = file_picker.value

    # Appeler la fonction select_images pour traiter les fichiers téléchargés
    select_images(files)

# Définir une fonction pour effacer la sortie
def clear_output_func(button):
    clear_output()

# Créer un widget pour sélectionner plusieurs fichiers
file_picker = widgets.FileUpload(multiple=True)

# Créer un bouton pour déclencher les prédictions
predict_button = widgets.Button(description="Predict")

# Créer un bouton pour effacer la sortie
clear_button = widgets.Button(description="Clear")

# Afficher les widgets
display(file_picker)
display(predict_button)
display(clear_button)

# Associer la fonction predict_images au bouton de prédiction
predict_button.on_click(predict_images)

# Associer la fonction clear_output_func au bouton d'effacement de la sortie
clear_button.on_click(clear_output_func)
```

On commence par charger un modèle pré-entraîné à partir d'un fichier h5 en utilisant la fonction `keras.models.load_model`. Ensuite, on définit une fonction, `select_images`, qui sera appelée pour traiter les images téléchargées. Cette fonction prend en paramètre les fichiers téléchargés et les parcourt un par un.

Pour chaque fichier, on convertit son contenu en un objet BytesIO, ce qui permet de manipuler l'image à l'aide de bibliothèques comme PIL ou Matplotlib. Ensuite, on convertit l'image en un tableau NumPy et on normalise les valeurs des pixels pour les rendre compatibles avec le modèle. Après la préparation de l'image, on fait une prédiction en utilisant le modèle chargé.

En fonction de la confiance de la prédiction, on attribue une étiquette et une couleur à l'image. Si la confiance de prédiction pour la classe "fake" est supérieure ou égale à 0.50, l'image est étiquetée comme "Fake" avec la confiance de prédiction affichée en rouge. Sinon, elle est étiquetée comme "Real" avec la confiance de prédiction pour la classe "real" affichée en bleu.

Enfin, on affiche l'image avec son étiquette prédite à l'aide de Matplotlib.

On définit également une fonction `predict_images` pour déclencher les prédictions lorsque l'utilisateur clique sur un bouton, et une fonction `clear_output_func` pour effacer la sortie précédente. Ces fonctions sont ensuite associées aux actions des boutons respectifs.

Enfin, des widgets sont créés pour permettre à l'utilisateur de sélectionner plusieurs fichiers et de déclencher les prédictions et le nettoyage de la sortie. Ces widgets sont ensuite affichés à l'écran.
