On définit une séquence de transformations pour préparer les images avant de les fournir au modèle de deep learning. On utilise la méthode `transforms.Compose` pour créer une chaîne de transformations qui seront appliquées séquentiellement à chaque image.

Les transformations sont les suivantes :

1. Redimensionner les images en 224x224 pixels en utilisant `transforms.Resize((224, 224))`. Cela assure que toutes les images ont la même taille, ce qui est souvent nécessaire pour l'entraînement des modèles de deep learning.

3. Normaliser les images en utilisant `transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])`. Cette transformation applique une normalisation aux valeurs de pixel des images en soustrayant la moyenne et en divisant par l'écart type pour chaque canal couleur (rouge, vert, bleu). Cette normalisation aide souvent à stabiliser et à accélérer la convergence de l'entraînement des modèles.

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(64 * 56 * 56, 512)
        self.fc2 = nn.Linear(512, 2)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64 * 56 * 56)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = SimpleCNN()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

On définit une classe `SimpleCNN` qui hérite de `nn.Module`, ce qui permet de créer un modèle de réseau de neurones convolutif (CNN) simple.

Dans la méthode `__init__`, on définit les différentes couches du réseau :
- `conv1` : Une couche de convolution avec 3 canaux d'entrée, 32 filtres, une taille de noyau de 3x3, un pas (stride) de 1 et un bourrage (padding) de 1.
- `pool` : Une couche de max-pooling avec une taille de noyau de 2x2 et un pas de 2.
- `conv2` : Une deuxième couche de convolution avec 32 canaux d'entrée, 64 filtres, une taille de noyau de 3x3, un pas de 1 et un bourrage de 1.
- `fc1` : Une couche entièrement connectée avec 64 * 56 * 56 neurones en entrée (la taille des tenseurs en sortie de la deuxième couche de convolution), et 512 neurones en sortie.
- `fc2` : Une autre couche entièrement connectée avec 512 neurones en entrée et 2 neurones en sortie (correspondant aux classes à prédire).

Dans la méthode `forward`, on définit la propagation avant (forward pass) du modèle :
- On applique d'abord la première couche de convolution suivie d'une fonction d'activation ReLU et d'une opération de max-pooling.
- Ensuite, on applique la deuxième couche de convolution avec une autre fonction ReLU et une autre opération de max-pooling.
- On redimensionne ensuite les sorties de la dernière couche de convolution pour les adapter à la forme attendue par la première couche entièrement connectée.
- On applique la première couche entièrement connectée suivie d'une fonction d'activation ReLU.
- Enfin, on applique la deuxième couche entièrement connectée pour obtenir les scores de sortie.

On instancie ensuite un objet de la classe `SimpleCNN` et on le déplace sur le GPU s'il est disponible. On définit également une fonction de perte (loss) de classification croisée (CrossEntropyLoss) et un optimiseur Adam pour entraîner le modèle.

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0
        for inputs, labels in tqdm(train_loader):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_losses.append(running_loss / len(train_loader))
        train_accuracies.append(100 * correct / total)

        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        val_losses.append(val_loss / len(val_loader))
        val_accuracies.append(100 * correct / total)

        print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {running_loss/len(train_loader):.4f}, Train Accuracy: {train_accuracies[-1]:.2f}%, Val Loss: {val_loss/len(val_loader):.4f}, Val Accuracy: {val_accuracies[-1]:.2f}%")

    return train_losses, val_losses, train_accuracies, val_accuracies
```

Cette fonction `train_model` est utilisée pour entraîner le modèle de classification sur les données d'entraînement et évaluer ses performances sur les données de validation. Voici ce qu'elle fait :

1. Initialise des listes vides pour stocker les pertes et les précisions d'entraînement et de validation à chaque époque.
2. Pour chaque époque :
   - Met le modèle en mode d'entraînement.
   - Parcourt les données d'entraînement en utilisant la boucle `for` pour les mini-lots (batches) :
     - Transfère les données sur le GPU si disponible.
     - Réinitialise les gradients du modèle.
     - Effectue une passe avant (forward pass) pour obtenir les prédictions.
     - Calcule la perte sur les prédictions et les étiquettes réelles.
     - Effectue une passe arrière (backward pass) pour calculer les gradients.
     - Met à jour les poids du modèle en utilisant l'optimiseur.
     - Calcule le nombre de prédictions correctes pour calculer la précision d'entraînement.
   - Calcule et stocke la perte et la précision moyennes d'entraînement pour cette époque.
   - Met le modèle en mode d'évaluation.
   - Évalue le modèle sur les données de validation sans calculer les gradients :
     - Calcule la perte moyenne sur les données de validation.
     - Calcule le nombre de prédictions correctes pour calculer la précision de validation.
   - Calcule et stocke la perte et la précision moyennes de validation pour cette époque.
   - Affiche les statistiques d'entraînement et de validation pour cette époque.
3. Retourne les listes de pertes et de précisions d'entraînement et de validation pour toutes les époques.

Cette fonction permet de suivre les performances du modèle pendant l'entraînement et d'identifier tout surajustement (overfitting) ou sous-ajustement (underfitting) potentiel. Elle est utile pour ajuster les hyperparamètres du modèle et de l'optimiseur.

```python
def calculate_metrics(model, val_loader):
    model.eval()
    true_labels = []
    pred_labels = []
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            true_labels.extend(labels.cpu().numpy())
            pred_labels.extend(predicted.cpu().numpy())

    cm = confusion_matrix(true_labels, pred_labels)
    print("Confusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(classification_report(true_labels, pred_labels, target_names=['Real', 'Fake']))
    
    accuracy = accuracy_score(true_labels, pred_labels)
    precision = precision_score(true_labels, pred_labels, average='weighted')
    recall = recall_score(true_labels, pred_labels, average='weighted')
    f1 = f1_score(true_labels, pred_labels, average='weighted')
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    
    return cm, accuracy, precision, recall, f1

cm, accuracy, precision, recall, f1 = calculate_metrics(model, val_loader)
```

On définit une fonction `calculate_metrics` pour évaluer les performances du modèle sur l'ensemble de validation. Voici ce qu'elle fait :

1. Met le modèle en mode d'évaluation en appelant `model.eval()`.
2. Initialise deux listes vides pour stocker les vraies étiquettes (`true_labels`) et les étiquettes prédites (`pred_labels`).
3. Parcourt les données de validation en utilisant une boucle `for` sur le chargeur de données de validation (`val_loader`).
4. Pour chaque mini-lot de données de validation :
   - Transfère les données sur le périphérique spécifié (GPU ou CPU).
   - Fait une prédiction en utilisant le modèle.
   - Récupère les étiquettes prédites en prenant l'indice de la classe prédite avec la plus haute probabilité.
   - Étend les listes `true_labels` et `pred_labels` avec les vraies et les étiquettes prédites, respectivement.
5. Calcule la matrice de confusion en utilisant les vraies et les étiquettes prédites.
6. Affiche la matrice de confusion et le rapport de classification.
7. Calcule et affiche la précision, le rappel, le score F1 et l'exactitude en utilisant les fonctions d'évaluation fournies par scikit-learn.
8. Retourne la matrice de confusion ainsi que les mesures d'évaluation (exactitude, précision, rappel, score F1).

 utile pour évaluer les performances globales du modèle sur un ensemble de validation et comprendre ses forces et ses faiblesses dans la classification des données réelles et falsifiées.

```python
import io
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing import image
import ipywidgets as widgets
from IPython.display import display, clear_output

# Charger le modèle pré-entraîné à partir du fichier h5
model = keras.models.load_model('models/IMAGE_DEEP_FAKE_MODEL-2.h5')

# Définir une fonction pour traiter la sélection des images téléchargées
def select_images(files):
    # Traiter chaque fichier téléchargé
    for file in files:
        # Convertir le fichier téléchargé en un objet BytesIO
        img = image.load_img(io.BytesIO(file['content']), target_size=(224, 224))  # Redimensionner à 224x224

        # Convertir l'image en un tableau NumPy et normaliser les valeurs de pixels
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x /= 255.0

        # Faire une prédiction en utilisant le modèle
        y_pred = model.predict(x)

        # Extraire le score de confiance pour la classe prédite
        confidence_score = y_pred[0][1]  # On suppose que le modèle renvoie la probabilité pour la classe '1' (faux)

        # Déterminer l'étiquette et la couleur de l'image en fonction du score de confiance
        if confidence_score >= 0.52:
            label = f'Fake avec un score de confiance de {confidence_score:.2f}'
            color = 'red'
        else:
            label = f'Real avec un score de confiance de {1 - confidence_score:.2f}'  # Le score de confiance pour le réel est 1 - le score pour le faux
            color = 'blue'

        # Afficher l'image avec l'étiquette
        plt.figure()
        plt.imshow(image.load_img(io.BytesIO(file['content'])))
        plt.title(label, color=color)
        plt.axis('off')
        plt.show()

# Définir une fonction pour déclencher les prédictions
def predict_images(button):
    # Obtenir les fichiers téléchargés à partir du widget de sélection de fichiers
    files = file_picker.value

    # Appeler la fonction select_images pour traiter les fichiers téléchargés
    select_images(files)

# Définir une fonction pour effacer la sortie
def clear_output_func(button):
    clear_output()

# Créer un widget pour sélectionner plusieurs fichiers
file_picker = widgets.FileUpload(multiple=True)

# Créer un bouton pour déclencher les prédictions
predict_button = widgets.Button(description="Predict")

# Créer un bouton pour effacer la sortie
clear_button = widgets.Button(description="Clear")

# Afficher les widgets
display(file_picker)
display(predict_button)
display(clear_button)

# Associer la fonction predict_images au bouton de prédiction
predict_button.on_click(predict_images)

# Associer la fonction clear_output_func au bouton d'effacement de la sortie
clear_button.on_click(clear_output_func)
```
